```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
x = 5*np.random.rand(100, 1)
y = 10*np.random.rand(100, 1)
```

```python
plt.scatter(x, y)
plt.title('Initial X and Y Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
y = 0.5*x+np.random.rand(100, 1)

plt.scatter(x, y)
plt.title('Revised X and Y Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
def bestFitLine(x, y, title, xLabel, yLabel):

  m = ((np.mean(x) * np.mean(y)) - np.mean(x * y)) / ((np.mean(x)**2) - np.mean(x**2))

  c = np.mean(y) - m * np.mean(x)
  
  plt.scatter(x, y)
  plt.plot([min(x), max(x)], [(min(x)*m+c), (max(x)*m+c)], color='red')
  plt.title(str(title))
  plt.xlabel(str(xLabel))
  plt.ylabel(str(yLabel))
  plt.show()
```

```python
bestFitLine(x, y, 'X and Y Scatter Plot with Best Fit Line', 'X', 'Y')
```
